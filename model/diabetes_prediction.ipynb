```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# from sklearn.ensemble import RandomForestClassifier
# from sklearn.ensemble import AdaBoostClassifier
# from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC

# import tensorflow as tf
# import keras
# from keras import layers
```

Load Data and process with One_Hot encodding


```python
df=pd.read_csv('../data/diabetes_thousend_data.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().values.any()
```

```python
y=df[['diabetes']]
y.head()
```

```python
x1=df.iloc[:,:-1]
x1.head()
```

```python
dummies1=pd.get_dummies(x1['gender'])
dummies2=pd.get_dummies(x1['smoking_history'])
mereged=pd.concat([x1,dummies1,dummies2],axis='columns')
x=mereged.drop(['gender','smoking_history'],axis='columns')
x.head()
```

```python
xtrain, xtest, ytrain, ytest=train_test_split(x, y, test_size=0.2, random_state=42)
```

Model creation

```python
model= SVC(kernel='linear') #SVM
```

```python
model.fit(xtrain.values,ytrain.values.ravel())
```

```python
weights = model.coef_
bias = model.intercept_

# Print the weights for each feature and the bias term
print("Weights (Coefficients):", weights)
print("Bias (Intercept):", bias)
```

```python
model.score(xtest.values,ytest.values)
```

```python
result=model.predict([[51.0,0,0,39.57,6.2,240,1,0,0,0,0,0,0,1,0]])
# result=model.predict([[73.0,0,0,26.62,5.0,155,1,0,0,1,0,0,0,0,0]])
print(result)
if result[0]==1:
    print("You have daibetes")
else:
    print("You don't have daibetes")

```
